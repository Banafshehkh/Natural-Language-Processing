```python
import os
from collections import Counter

def extract_word_frequency(filenames):
    word_counter = [Counter() for _ in range(len(filenames))]
    for i, filename in enumerate(filenames):
        with open(filename, 'r') as file:
            for line in file:
                line = line.lower()
               
                words = line.strip().split()
                word_counter[i].update(words)

    return word_counter

def get_frequent_words(word_counter, threshold):
    word_frequency = Counter()
    for counter in word_counter:
        for word, count in counter.items():
            word_frequency[word] += 1
    frequent_words = []
    total_files = len(word_counter)
    for word, count in word_frequency.items():
        if count >= threshold * total_files:
            frequent_words.append(word)
    return frequent_words

filenames = [f"/content/exercize1/text_{i}.txt" for i in range(1, 11)]
word_counter = extract_word_frequency(filenames)
frequent_words = get_frequent_words(word_counter, 0.8)

print("Frequent words:", frequent_words)
```
